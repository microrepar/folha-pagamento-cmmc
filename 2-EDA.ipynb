```python
def normaliza_cargo(cargo):
    if cargo.startswith('CHEFE DE DIV'):
        return 'CHEFE DE DIVISAO'
    
    if cargo.startswith('DIRETOR D'):
        return 'DIRETOR DE DEPARTAMENTO I'
    
    if cargo.startswith("PROF.'I' DE"):
        return 'PROFESSOR ENSINO FUNDAMENTAL I'
    if cargo.startswith("PROFESSOR I DE"):
        return 'PROFESSOR ENSINO FUNDAMENTAL I'
    if cargo.startswith('PROF.I" DE'):
        return 'PROFESSOR ENSINO FUNDAMENTAL I'    
    if cargo.startswith('PROF. I" DE'):
        return 'PROFESSOR ENSINO FUNDAMENTAL I'
    
    if cargo.startswith('PROFESSOR II DE'):
        return 'PROFESSOR ENSINO FUNDAMENTAL II'
    if cargo.startswith("PROF.'II' DE"):
        return 'PROFESSOR ENSINO FUNDAMENTAL II'
    if cargo.startswith('PROF.II" DE'):
        return 'PROFESSOR ENSINO FUNDAMENTAL II'
    if cargo.startswith("PROF.'II'"):
        return 'PROFESSOR ENSINO FUNDAMENTAL II'
    
    if cargo.startswith('PROFESSOR DE EDUCAÇÃO BÁSICA I'):
        return 'PROFESSOR EDUCACAO BASICA I'
    if cargo.startswith('PROF.EDUCACAO INFANTIL'):
        return 'PROFESSOR EDUCACAO BASICA I'
    if cargo.startswith('PROFESSOR EDUC INFANTIL-'):
        return 'PROFESSOR EDUCACAO BASICA I'
    if cargo.startswith('PROF. EDUC. INFANTIL'):
        return 'PROFESSOR EDUCACAO BASICA I'
    if cargo.startswith('PROF.EDUC.INFANTIL-J'):
        return 'PROFESSOR EDUCACAO BASICA I'
    if cargo.startswith('PROFESSOR EDUCAÇÃO BÁSICA II'):
        return 'PROFESSOR EDUCACAO BASICA II'
    if cargo.startswith('PROFESSOR EDUCACAO BASICA II'):
        return 'PROFESSOR EDUCACAO BASICA II'
       
    return cargo


df['cargo_normalizado'] = df['cargo']
df['cargo_normalizado'] = df['cargo_normalizado'].apply(normaliza_cargo)
print(len(df['cargo'].unique()))
print(len(df['cargo_normalizado'].unique()))
df[['matricula', 'nome', 'cargo', 'cargo_normalizado']][df['cargo_normalizado'].str.contains(r'PROF', regex=True)]['cargo_normalizado'].unique()
```

```python
import plotly.graph_objects as go

import numpy as np
np.random.seed(1)

x = np.random.rand(100)
y = np.random.rand(100)

fig = go.FigureWidget([go.Scatter(x=x, y=y, mode='markers')])

scatter = fig.data[0]
colors = ['#a3a7e4'] * 100
scatter.marker.color = colors
scatter.marker.size = [10] * 100
fig.layout.hovermode = 'closest'


# create our callback function
def update_point(trace, points, selector):
    c = list(scatter.marker.color)
    s = list(scatter.marker.size)
    for i in points.point_inds:
        c[i] = '#bae2be'
        s[i] = 20
        with fig.batch_update():
            scatter.marker.color = c
            scatter.marker.size = s


scatter.on_click(update_point)

fig
```

```python
df = load_data()

def get_maiores_pagamentos_matriculas(df, col, n=10):
    idx_maiores = df.groupby('cargo_normalizado')[col].idxmax()
    return df.loc[idx_maiores, col]


df_result = df[df['ano'].isin(['2023']) & df['mes'].isin(['06'])].groupby('cargo_normalizado').apply(get_maiores_pagamentos_matriculas, 'valor_bruto')
print(df_result.sort_values(ascending=False).nlargest(10).index.get_level_values(1).tolist())
print(df_result.sort_values(ascending=False).nlargest(10).index.get_level_values(0).tolist())
print(df_result.sort_values(ascending=False).nlargest(10).values)
df[df['ano'].isin(['2023']) & df['mes'].isin(['06'])].loc[df_result.sort_values(ascending=False).nlargest(10).index.get_level_values(1).tolist()]['matricula']
```
